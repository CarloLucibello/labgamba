```python
# Carico le librerie matematiche

%pylab              
%matplotlib inline

# Definisco la funzione gaussiana

def gaussiana(x,xmedio,sigma):
    return exp( -( x - xmedio )**2 / (2*sigma**2) ) / sqrt( 2*pi*sigma**2 )
```

```python
# ne disegno il grafico con xmedio = 5 e sigma = 0.5, 1.0, 2.0

xmedio = 5
x      = linspace(0,10,200)
fmt    = r'$\sigma=%3.1f$'            # formato utilizzato per la legenda

sigma = 0.5;  plot(x,gaussiana(x,xmedio,sigma),label=fmt%sigma)
sigma = 1.0;  plot(x,gaussiana(x,xmedio,sigma),label=fmt%sigma)
sigma = 2.0;  plot(x,gaussiana(x,xmedio,sigma),label=fmt%sigma)

legend()
xlabel(r'$x$')
ylabel(r'$p(x)$')
grid()
```

```python
%pylab              
%matplotlib inline

from scipy import stats

x=linspace(-4,4,200)
fig=figure(figsize=(14,4))
fig.add_subplot(121)

plot(x,stats.norm.pdf(x))                # 'probability density function'

title(u'densità di probabilità normale standard')
xlabel(r'$z$',fontsize=20)
ylabel(r'$\varphi(z)$',fontsize=20)
ylim((0,1))
grid()
fig.add_subplot(122)

plot(x,stats.norm.cdf(x))                # 'cumulative distribution function'

title(u'distribuzione di probabilità cumulata normale standard')
xlabel(r'$z$',fontsize=20)
ylabel(r'$\Phi(z)$',fontsize=20)
grid()
```
