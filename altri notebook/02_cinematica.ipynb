```python
# Carico le librerie matematiche

%pylab
%matplotlib inline

# MOTO UNIFORMEMENTE ACCELERATO

T       = 1.2                   # s        tempo finale  
dt      = 0.1                   # s        passo temporale
N       = int(round(T/dt))      #          numero di istanti temporali considerati
t       = arange(0,T,dt)        #          creo un vettore con valori del tempo crescenti spaziati di dt

# Accelerazione costante:

a0      = 1.                    # m/s**2

# Creo dei vettori contenenti N zeri, che saranno riempiti con i valori di x,v,a:

x       = zeros(N)              # m
v       = zeros(N)              # m/s
a       = zeros(N) + a0         # m/s**2   qui invece di N zeri metto N valori uguali ad a0

# Dati iniziali

x[0]    = 0.5                   # m        posizione iniziale
v[0]    = 0                     # m/s      velocità iniziale

# Integrazione numerica dell'equazione differenziale:

for i in range(0,N-1):          
    v[i+1] = v[i] + a[i] * dt
    x[i+1] = x[i] + v[i] * dt
       
#print t                         # s        tempi
#print a                         # m/s**2   accelerazioni
#print a*dt                      # m/s      incrementi di velocità
#print v                         # m/s      velocità
#print v*dt                      # m        incrementi di posizione
print x                         # m        posizione
```

```python
fig=figure(figsize=(5,7))

fig.add_subplot(311)
plot(t,x,'-bo')
grid()
ylabel('Posizione x(t) [m]')

fig.add_subplot(312)
plot(t,v,'-ro')
grid()
ylabel('Velocita'' v(t) [m/s]')

fig.add_subplot(313)
plot(t,a,'-go')
grid()
xlabel('Tempo t [s]')
ylabel('Accelerazione a(t) [m/s**2]')

# ESERCIZIO: confrontare sullo stesso grafico la soluzione numerica appena ottenuta e 
# la legge esatta del moto uniformemente accelerato vista a lezione.
# Provare a variare il passo temporale dt, prendendolo prima molto grande, poi più piccolo.
# La precisione della soluzione numerica deve risultare tanto più grande quanto più dt è piccolo.
```
