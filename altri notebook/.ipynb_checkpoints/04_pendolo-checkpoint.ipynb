```python
%pylab inline                   

from scipy.special import ellipk                    # importo l'integrale ellittico completo del primo tipo K(m)
```

```python
# Periodo del pendolo in funzione dell'ampiezza delle oscillazioni

L     = 1                                           # m
g     = 9.81                                        # m/s2

theta = linspace(-pi,pi,1000)

T0    = 2*pi*sqrt(L/g)                              # periodo delle piccole oscillazioni
T1    = T0*(1+theta**2/16.)                         # periodo con correzione del second'ordine in theta
T     = 4*sqrt(L/g)*ellipk(sin(theta/2.)**2)        # periodo del pendolo

fig=figure(figsize=(10,6))
ax=fig.add_subplot(111)

plot(theta,T/T0,'b',linewidth=3,label='periodo esatto')
plot(theta,T1/T0,'y',linewidth=3,label="second'ordine")
plot(theta,0*theta+1,'r',linewidth=3,label='piccole oscillazioni')

eps=0.126                                           # regione nella quale l'approssimazione di piccole oscillazioni
plot(-eps+0*theta,theta,'g')                        # da un errore relativo inferiore a 0.001
plot( eps+0*theta,theta,'g')

ylim(0,4)
ax.set_xticks([-pi,-pi/2,0,pi/2,pi])
ax.set_xticklabels(['$-\pi$',r'$-\frac{\pi}{2}$','$0$',r'$\frac{\pi}{2}$','$\pi$'],fontsize=20)
ax.set_yticks(arange(5))
ax.set_yticklabels(['$0$','$1$','$2$','$3$','$4$'],fontsize=20)
xlabel('ampiezza delle oscillazioni',fontsize=16)
ylabel('periodo esatto / piccole oscillazioni',fontsize=16)
legend(fontsize=16)
grid()
```
