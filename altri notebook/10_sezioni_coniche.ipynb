```python
# Sezioni coniche

%pylab inline

d=1.                                                      # (m) distanza del fuoco dalla direttrice
L=3                                                       # (m) fattore di scala
NP=60                                                     # numero di punti

def r(theta):                                             # Equazione della sezione conica in coordinate polari
   return eps*d/(1.-eps*cos(theta))

fig=figure(figsize=(18,10))

theta=linspace(0,2*pi,NP)

eps=0.3                                                   # ellisse
plot(r(theta)*cos(theta),r(theta)*sin(theta),'.-',label='eps = %.2f (ellisse)'%eps)

eps=0.8                                                   # ellisse
plot(r(theta)*cos(theta),r(theta)*sin(theta),'.-',label='eps = %.2f (ellisse)'%eps)

eps=0.89                                                  # ellisse
plot(r(theta)*cos(theta),r(theta)*sin(theta),'.-',label='eps = %.2f (ellisse)'%eps)

eps=1                                                     # parabola
plot(r(theta)*cos(theta),r(theta)*sin(theta),'.-',label='eps = %.2f (parabola)'%eps)

eps=1.5                                                   # iperbole
maxtheta=arccos(1/eps)                                    # (voglio disegnare un singolo ramo dell'iperbole, quindi
theta=linspace(maxtheta,2*pi-maxtheta,NP)                 #  mi restringo ai valori di theta per i quali r > 0)   
plot(r(theta)*cos(theta),r(theta)*sin(theta),'.-',label='eps = %.2f (iperbole)'%eps)      

plot(0,0,'o',markersize=10)                               # fuoco
y=linspace(-L,L,200) 
plot(-d+0*y,y,'.-',label='direttrice')                    # direttrice

axis('scaled')                                            # stesse unità di misura lungo l'asse x e l'asse y
axis([-d-1,3*L-d,-L,L])
legend()
grid()
```
